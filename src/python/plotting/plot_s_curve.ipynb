# Plot s-curve

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns; sns.set()
```

### Plotting

```python
widthMM = 170 
widthInch = widthMM / 25.4
ratio = 0.66666
heigthInch = ratio*widthInch

SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
sns.set_style("ticks")

dpi = 300
```

### S-curve of augmented LSH

```python
def AND_OR(p,n,m):
    # n times AND
    # m times OR
    
    return 1.0-np.power((1.0-np.power(p,n)),m)
```

```python
p = np.linspace(0,1,num=1000)
```

```python
s = np.linspace(0,1.0,num=1000)
inverseAbs = 2.0*s -1
pSim = 1.0 - (np.arccos(s))/(np.pi)
```

```python
fig, ax = plt.subplots(figsize=(widthInch, heigthInch), dpi=dpi, facecolor='w', edgecolor='k')
plt.xlabel('$s(x_{i},x_{j})$')
plt.ylabel('$P_{n,m}([x_{i},x_{j}])$')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#ax.tick_params(color='gray', labelcolor='black')
#for spine in ax.spines.values():
#    spine.set_edgecolor('gray')

plt.xlim([0.55,1])

n = 16
m = 32
y = AND_OR(pSim,n,m)
plt.plot(s,y,label="m = {:2d} n = {:2d} ".format(m,n))

n = 32
m = 32
y = AND_OR(pSim,n,m)
plt.plot(s,y,label="m = {:2d} n = {:2d} ".format(m,n))

n = 32
m = 64
y = AND_OR(pSim,n,m)
plt.plot(s,y,label="m = {:2d} n = {:2d} ".format(m,n))

plt.legend(loc='best',frameon=False)

plt.tight_layout()
plt.savefig("AugmentedLSH.pdf")
plt.show()
```
